```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
import joblib
```

```python
model = load_model("model.h5")
onehot_geo = joblib.load("onehot_geo.pkl")
label_encoder_gender=joblib.load("label_encoder_gender.pkl")
scaler=joblib.load("scaler.pkl")
```

```python
input_data = {
    'CreditScore': 600,
    'Geography': 'France',
    'Gender': 'Male',
    'Age': 40,
    'Tenure': 3,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 50000
}

df=pd.DataFrame([input_data])
```

```python
encoded_geo = onehot_geo.transform([[df['Geography'].iloc[0]]])[0]
encoded_geo_df=pd.DataFrame([encoded_geo],columns=onehot_geo.get_feature_names_out(['Geography']))
df=pd.concat([df.drop('Geography',axis=1),encoded_geo_df],axis=1)
df['Gender']=label_encoder_gender.transform([df['Gender'].iloc[0]])[0]
df.head()
```

```python
input_scaled=scaler.transform(df)
input_scaled
```

```python
prediction=model.predict(input_scaled)
prediction_proba=prediction[0][0]
prediction_proba
```

```python
if prediction_proba > 0.5:
    print("Customer like to Chrun")
else:
    print("Customer not likely to Churn")
```
